```python
import pandas as pd
iris = pd.read_csv("C:/Users/MSCIT/Downloads/iris (1).csv")
print(iris.head())
```

```python
# get columns to plot
columns = iris.columns.drop(['variety'])
# create x data
x_data = range(0, iris.shape[0])
# create figure and axis
fig, ax = plt.subplots()
for column in columns:
    ax.plot(x_data, iris[column], label=column)
    # set title and legend
    ax.set_title('Iris Dataset')
    ax.legend()
```

```python
#Bar chart
wine_reviews=pd.read_csv("C:/Users/MSCIT/Downloads/winemag-data-130k-v2.csv")
wine_reviews.head()
```

```python
# Bar chart
# create a figure and axis 
fig, ax=plt.subplots()
# count the occurrence of each class
data = wine_reviews['points'].value_counts()
# get x and y data
points = data.index
frequency = data.values
ax.bar(points,frequency)
ax.set_title('wine reviews scores')
ax.set_xlabel('points')
ax.set_ylabel('Frequency')
```

```python
import numpy as np
import matplotlib.pyplot as plt
objects = ('Python', 'C++', 'Java', 'Perl','Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
# Bar Chart
# Axis positions as first parameter list,it can be floating point
# Y Values as 2nd parameter list
# Alpha is transparency,
# Align can be center or edge
# Color can be single value or a list of color codes, one for each
plt.bar(y_pos, performance, width=0.5, align='center', alpha=0.5,color=['r','r','g','g','b','b'])
# to define label for x axis values
plt.xticks(y_pos,objects)
plt.xlabel('Programing languages')
plt.ylabel('Performance level')
plt.show()
        
        
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=[15,10])
totalDeath = [113055,37312,5971,7473,33964]
totalRecovery = [773480,325602,230688,129095,166584]
activeCases = [1139958,347973,239999,129360,34730]
country = ['USA', 'BRAZIL', 'RUSSIA', 'SPAIN', 'UK', 'INDIA']
x=np.arange(len(totalDeath))
plt.bar(x, totalDeath, color = 'black',width = 0.25)
plt.bar(x + 0.25, totalRecovery, color = 'g', width =0.25)
plt.bar(x + 0.5, activeCases,color = 'b' , width = 0.25)
plt.xticks([i+0.25 for i in range(5)],country)
plt.title("Bar representing the total deaths, total recover cases and active cases country wise")
plt.xlabel('Countries')
plt.ylabel('Cases')
plt.savefig('4BarPlot.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a jupyter notebook, include:
%matplotlib inline
#Pie chart,where the slices will be ordered and plotted counter-clockwise:
labels = ['Civil', 'Electrical', 'Mechanical', 'Chemical']
sizes = [15,50,45,10]
fig, ax= plt.subplots()
ax.pie(sizes, labels=labels,autopct='%1.1f%%')
ax.axis('equal') # equal aspectr ratio ensures the pie chart is circular
ax.set_title('Engineering Diciplines')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# if using a jupyter notebook, include:
%matplotlib inline
#Pie chart,where the slices will be ordered and plotted counter-clockwise:
labels = ['Civil', 'Electrical', 'Mechanical', 'Chemical']
sizes = [15,50,45,10]
# explode out the 'Chemical' pie piece by offsetting it a greater amount
explode = (0.1,0.1,0.1,0.4)
fig,ax=plt.subplots()
ax.pie(sizes,
      explode=explode,
      labels=labels,
      autopct='%1.1f%%',
      shadow=True,
      startangle=90)
ax.axis('equal')
ax.set_title('Engineering Diciplines')
plt.show()
```
